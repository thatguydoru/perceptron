```python
from perceptron import Perceptron

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def gen_hmap(scale, model):
    inputs = []
    for x in range(scale+1):
        for y in range(scale+1):
            inputs.append([x / scale, y / scale])

    preds = [i for i in model.predict(inputs)]

    hmap = []
    for x in range(scale+1):
        hmap.append([])
    for idx, h in enumerate(hmap):
        slc = preds[(scale+1)*idx:(scale+1)*(idx+1)]
        for p in slc:
            h.append(p)

    labels = []
    for n in inputs:
        if n[0] not in labels:
            labels.append(n[0])

    df = pd.DataFrame(hmap, index=labels, columns=labels)    
    sns.heatmap(df)
```

```python
# AND data
train_features = [[0, 0], [0, 1], [1, 0], [1, 1]]
train_targets = [0, 1, 1, 1]

ptron = Perceptron(2)
ptron.train(train_features, train_targets, 500)

gen_hmap(1, ptron)
```
